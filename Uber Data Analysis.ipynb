```python
# #  What is the most popular starting point for the Uber drivers

df_start=pd.DataFrame(df1["START*"].value_counts())

df_start= df_start.reset_index()
df_start=df_start.rename(columns = {"index":"Start","START*":"Count"})
```

```python
df_start.loc[df_start["Count"]==max(df_start["Count"])]
```

```python
# What is the most popular dropping point for the Uber drivers

df_stop= pd.DataFrame(df1["STOP*"].value_counts())
df_stop=df_stop.reset_index()
df_stop=df_stop.rename(columns={"index":"STOP","STOP*":"Count"})

df_stop[df_stop["Count"]==max(df_stop["Count"])]
```

```python
# List the most frequent route taken by Uber drivers
df_group = pd.DataFrame(df1.groupby(["START*","STOP*"]).size())
df_group.rename(columns = {0:"count"},inplace = True)
df_group=df_group.sort_values(["count"],ascending=False)
df_group[df_group["count"]==max(df_group["count"])]
```

```python
#  Print all types of purposes for the trip in an array
print(np.array(df["PURPOSE*"].unique()))
df3=pd.DataFrame(df1["MILES*"].groupby(df1["PURPOSE*"]).sum())
df3.reset_index(inplace= True)
df3
```

```python
# Plot a bar graph of Purposes vs Distance
df3.plot( kind='bar',x="PURPOSE*",y="MILES*")
```
